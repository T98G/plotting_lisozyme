```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("gyrate.xvg", comments = ["#", "@"])
```

```python
np.shape(data)
```

```python
times = data[:, 0]
```

```python
radious = data[:, 1:]
```

```python
radious.shape
```

```python
mean_radious = np.mean(radious, axis = 1)
```

```python
mean_radious.shape
```

```python
times.shape
```

```python
fig, ax = plt. subplots()
ax.plot(times, mean_radious, "g-")
ax.set(xlabel = "Time (ns)", ylabel = "Rg (nm)", 
       title = "Radious of Gyration\n1AKI Unrestreined MD", ylim = [1.2, 1.25])
fig.savefig("Radious of Gyration.pdf")
plt.show()
```
